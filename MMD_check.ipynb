```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_svmlight_file

train = load_svmlight_file('data/mmd_mnist_torch')
x, y = train
x = x.todense()

sortind = np.argsort(y)
x = x[sortind, :]
y = y[sortind]

# test = load_svmlight_file('data/mmd_mnist_torch.t')
# testx, testy = test
# testx = testx.todense()
```

```python
import random
pselected = np.load('protos/protos_1500.npy')
cselected = np.load('critics/crits_1500.npy')

for i in range(10):
    data = y[cselected]
    ins = np.where(data == i)[0]
    ins = random.choice(ins)
    print(ins, y[cselected[ins]])

    plt.figure()
    plt.imshow(x[cselected[ins], :].reshape(28, 28))
    plt.savefig('images/crits_{}.png'.format(y[cselected[ins]]))
plt.show()
```
